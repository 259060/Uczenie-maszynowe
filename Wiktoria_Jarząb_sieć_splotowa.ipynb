Kod klasyfikuje dane obrazów ze zbioru danych CIFAR10. Możliwe klasy:

* 0	- samolot,
* 1	- samochód,
* 2	- ptak,
* 3	- kot,
* 4	- jeleń,
* 5	- pies,
* 6	- żaba,
* 7	- koń,
* 8	- statek,
* 9	- ciężarówka.

Obrazy są kolorowe (RGB) i mają rozmiar 32x32. Zbiór uczący składa się z 50 000, a testowy z 10 000. Wraz z tworzeniem każdego modelu, wartości danych będą dzielone przez 255 (maksymalna wartość RGB), w celu zmniejszenia zakresu do [0,1].

```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
import numpy as np

(x_train, y_train), (x_test, y_test) = keras.datasets.cifar10.load_data()
```

Porównujemy dwa rodzaje modelu. Jeden składający się z dwóch warst splotowych i drugi składający się z trzech warstw splotowych. Po porównaniu wyników okazuje się, że model trójwarstwowy osiaga lepszy wynik.

```python
def tworzenie_modelu_2_warstwy():
  model = tf.keras.Sequential([
      tf.keras.layers.Input(shape=(32, 32, 3)),
      tf.keras.layers.Rescaling(1./255),

      tf.keras.layers.Conv2D(32, kernel_size = 3),
      tf.keras.layers.Activation("relu"),
      tf.keras.layers.MaxPool2D(pool_size = 2, strides = 2),

      tf.keras.layers.Conv2D(64, kernel_size = 3),
      tf.keras.layers.Activation("relu"),
      tf.keras.layers.MaxPool2D(pool_size = 2, strides = 2),

      tf.keras.layers.Flatten(),
      tf.keras.layers.Dense(256),
      tf.keras.layers.Activation("relu"),
      tf.keras.layers.Dense(10),
      tf.keras.layers.Activation("softmax")
  ])
  return model
```

```python
def tworzenie_modelu_3_warstwy():
  model = tf.keras.Sequential([
      tf.keras.layers.Input(shape=(32, 32, 3)),
      tf.keras.layers.Rescaling(1./255),

      tf.keras.layers.Conv2D(32, kernel_size = 3),
      tf.keras.layers.Activation("relu"),
      tf.keras.layers.MaxPool2D(pool_size = 2, strides = 2),

      tf.keras.layers.Conv2D(64, kernel_size = 3),
      tf.keras.layers.Activation("relu"),
      tf.keras.layers.MaxPool2D(pool_size = 2, strides = 2),

      tf.keras.layers.Conv2D(128, kernel_size = 3),
      tf.keras.layers.Activation("relu"),
      tf.keras.layers.MaxPool2D(pool_size = 2, strides = 2),

      tf.keras.layers.Flatten(),
      tf.keras.layers.Dense(256),
      tf.keras.layers.Activation("relu"),
      tf.keras.layers.Dense(10),
      tf.keras.layers.Activation("softmax")
  ])
  return model
```

```python
def modelowanie(model, epoki=20):

    model.compile(
      optimizer=keras.optimizers.Adam(3e-4),
      loss=keras.losses.SparseCategoricalCrossentropy(),
      metrics=['accuracy']
    )

    history = model.fit(
        x_train, y_train,
        epochs=epoki,
        batch_size=64,
        validation_data=(x_test, y_test)
    )

    return history
```

```python
modelowanie(tworzenie_modelu_2_warstwy())
```

```python
modelowanie(tworzenie_modelu_3_warstwy())
```

Porównujemy trzy kolejne rodzaje modeli, w których dodajemy różne wielkości dropout. Korzystamy z trójwarstwowego modelu, który poprzednio okazał się skuteczniejszy. Po porównaniu wyników okazuje się, że dropout 0.15 jest najlepszy.

```python
def tworzenie_modelu_3_warstwy_dropout_15():
  model = tf.keras.Sequential([
      tf.keras.layers.Input(shape=(32, 32, 3)),
      tf.keras.layers.Rescaling(1./255),

      tf.keras.layers.Conv2D(32, kernel_size = 3),
      tf.keras.layers.Activation("relu"),
      tf.keras.layers.MaxPool2D(pool_size = 2, strides = 2),

      tf.keras.layers.Conv2D(64, kernel_size = 3),
      tf.keras.layers.Activation("relu"),
      tf.keras.layers.MaxPool2D(pool_size = 2, strides = 2),

      tf.keras.layers.Conv2D(128, kernel_size = 3),
      tf.keras.layers.Activation("relu"),
      tf.keras.layers.MaxPool2D(pool_size = 2, strides = 2),

      tf.keras.layers.Flatten(),
      tf.keras.layers.Dense(256),
      tf.keras.layers.Activation("relu"),
      tf.keras.layers.Dropout(0.15),
      tf.keras.layers.Dense(10),
      tf.keras.layers.Activation("softmax")
  ])
  return model
```

```python
def tworzenie_modelu_3_warstwy_dropout_25():
  model = tf.keras.Sequential([
      tf.keras.layers.Input(shape=(32, 32, 3)),
      tf.keras.layers.Rescaling(1./255),

      tf.keras.layers.Conv2D(32, kernel_size = 3),
      tf.keras.layers.Activation("relu"),
      tf.keras.layers.MaxPool2D(pool_size = 2, strides = 2),

      tf.keras.layers.Conv2D(64, kernel_size = 3),
      tf.keras.layers.Activation("relu"),
      tf.keras.layers.MaxPool2D(pool_size = 2, strides = 2),

      tf.keras.layers.Conv2D(128, kernel_size = 3),
      tf.keras.layers.Activation("relu"),
      tf.keras.layers.MaxPool2D(pool_size = 2, strides = 2),

      tf.keras.layers.Flatten(),
      tf.keras.layers.Dense(256),
      tf.keras.layers.Activation("relu"),
      tf.keras.layers.Dropout(0.25),
      tf.keras.layers.Dense(10),
      tf.keras.layers.Activation("softmax")
  ])
  return model
```

```python
def tworzenie_modelu_3_warstwy_dropout_50():
  model = tf.keras.Sequential([
      tf.keras.layers.Input(shape=(32, 32, 3)),
      tf.keras.layers.Rescaling(1./255),

      tf.keras.layers.Conv2D(32, kernel_size = 3),
      tf.keras.layers.Activation("relu"),
      tf.keras.layers.MaxPool2D(pool_size = 2, strides = 2),

      tf.keras.layers.Conv2D(64, kernel_size = 3),
      tf.keras.layers.Activation("relu"),
      tf.keras.layers.MaxPool2D(pool_size = 2, strides = 2),

      tf.keras.layers.Conv2D(128, kernel_size = 3),
      tf.keras.layers.Activation("relu"),
      tf.keras.layers.MaxPool2D(pool_size = 2, strides = 2),

      tf.keras.layers.Flatten(),
      tf.keras.layers.Dense(256),
      tf.keras.layers.Activation("relu"),
      tf.keras.layers.Dropout(0.50),
      tf.keras.layers.Dense(10),
      tf.keras.layers.Activation("softmax")
  ])
  return model
```

Sprawdzamy teraz zachowanie modelu w przypadku dodania batch normalization. Jednakże wynik walidacji w tym przypadku się pogarsza.

```python
def tworzenie_modelu_3_warstwy_dropout_25_batch_normalization():
  model = tf.keras.Sequential([
      tf.keras.layers.Input(shape=(32, 32, 3)),
      tf.keras.layers.Rescaling(1./255),

      tf.keras.layers.Conv2D(32, kernel_size = 3),
      tf.keras.layers.BatchNormalization(),
      tf.keras.layers.Activation("relu"),
      tf.keras.layers.MaxPool2D(pool_size = 2, strides = 2),

      tf.keras.layers.Conv2D(64, kernel_size = 3),
      tf.keras.layers.BatchNormalization(),
      tf.keras.layers.Activation("relu"),
      tf.keras.layers.MaxPool2D(pool_size = 2, strides = 2),

      tf.keras.layers.Conv2D(128, kernel_size = 3),
      tf.keras.layers.BatchNormalization(),
      tf.keras.layers.Activation("relu"),
      tf.keras.layers.MaxPool2D(pool_size = 2, strides = 2),

      tf.keras.layers.Flatten(),
      tf.keras.layers.Dense(256),
      tf.keras.layers.Activation("relu"),
      tf.keras.layers.Dropout(0.25),
      tf.keras.layers.Dense(10),
      tf.keras.layers.Activation("softmax")
  ])
  return model
```

```python
modelowanie(tworzenie_modelu_3_warstwy_dropout_25_batch_normalization())
```

Jako ostatnie sprawdzamy czy dodanie padding do Conv2D zmieni wynik i owszem walidacja osiąga w takim przypadku najlepszy wynik.

```python
def tworzenie_modelu_3_warstwy_dropout_25_padding():
  model = tf.keras.Sequential([
      tf.keras.layers.Input(shape=(32, 32, 3)),
      tf.keras.layers.Rescaling(1./255),

      tf.keras.layers.Conv2D(32, kernel_size = 3, padding='same'),
      tf.keras.layers.BatchNormalization(),
      tf.keras.layers.Activation("relu"),
      tf.keras.layers.MaxPool2D(pool_size = 2, strides = 2),

      tf.keras.layers.Conv2D(64, kernel_size = 3, padding='same'),
      tf.keras.layers.BatchNormalization(),
      tf.keras.layers.Activation("relu"),
      tf.keras.layers.MaxPool2D(pool_size = 2, strides = 2),

      tf.keras.layers.Conv2D(128, kernel_size = 3, padding='same'),
      tf.keras.layers.BatchNormalization(),
      tf.keras.layers.Activation("relu"),
      tf.keras.layers.MaxPool2D(pool_size = 2, strides = 2),

      tf.keras.layers.Flatten(),
      tf.keras.layers.Dense(256),
      tf.keras.layers.Activation("relu"),
      tf.keras.layers.Dropout(0.25),
      tf.keras.layers.Dense(10),
      tf.keras.layers.Activation("softmax")
  ])
  return model
```
