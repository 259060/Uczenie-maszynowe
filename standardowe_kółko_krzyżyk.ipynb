```python
kolumny = 3
wiersze = 3
koniec = False
remis = False
pozycja = False
znak_gracz = str()
tura = str()

def tworzenie_planszy():
  plansza = [[" "] * kolumny for _ in range(wiersze)]
  return plansza

def wyświetlenie_planszy(plansza):
  for i in range(wiersze):
    print("|", end ="")
    for j in range(kolumny):
      print (plansza[i][j], "|", end ="")
    print("")

def czy_poprawna_pozycja(plansza, pozycja_kolumna, pozycja_wiersz):
  if pozycja_kolumna >= 3 or pozycja_wiersz >= 3:
    return False
  if plansza[pozycja_wiersz][pozycja_kolumna] == " ":
    return True
  else:
    return False

def czy_remis(plansza):
  return all(pole != " " for wiersz in plansza for pole in wiersz)

def czy_koniec(plansza):
  for wiersz in plansza:
    if wiersz[0] == wiersz[1] == wiersz[2] != " ":
      return True

  for kolumna in range(kolumny):
    if plansza[0][kolumna] == plansza[1][kolumna] == plansza[2][kolumna] != " ":
      return True

  if plansza[0][0] == plansza[1][1] == plansza[2][2] != " ":
    return True
  if plansza[0][2] == plansza[1][1] == plansza[2][0] != " ":
    return True

    return False
```

```python
def ocena(plansza):
  if czy_koniec(plansza):
    # sprawdzamy kto wygrał. 1 jeżeli komputer, -1 jeżeli gracz, 0 jeżeli remis
    for wiersz in plansza:
      if wiersz[0] == wiersz[1] == wiersz[2] != " ":
        return 1 if wiersz[0] == znak_komputer else -1

    for kolumna in range(kolumny):
      if plansza[0][kolumna] == plansza[1][kolumna] == plansza[2][kolumna] != " ":
        return 1 if plansza[0][kolumna] == znak_komputer else -1

    if plansza[0][0] == plansza[1][1] == plansza[2][2] != " ":
      return 1 if plansza[0][0] == znak_komputer else -1

    if plansza[0][2] == plansza[1][1] == plansza[2][0] != " ":
      return 1 if plansza[0][2] == znak_komputer else -1

  return 0

def minmax(plansza, maksymalizacja = True):
  if czy_koniec(plansza) == True or czy_remis(plansza) == True:
    return ocena(plansza), None

  # maksymalizujemy wynik komputera
  if maksymalizacja == True:
    najlepsza_wartość = -10
    najlepszy_ruch = None

    for w in range(wiersze):
      for k in range(kolumny):
        if plansza[w][k] == " ":
          plansza[w][k] = znak_komputer
          wynik = minmax(plansza, False)[0] # zmiana na minimalizację dla gracza
          plansza[w][k] = " "

          if wynik > najlepsza_wartość:
            najlepsza_wartość = wynik
            najlepszy_ruch = (w, k)

    return najlepsza_wartość, najlepszy_ruch

  else:
    najlepsza_wartość = 10
    najlepszy_ruch = None

    for w in range(wiersze):
      for k in range(kolumny):
        if plansza[w][k] == " ":
          plansza[w][k] = znak_gracz
          wynik = minmax(plansza, True)[0] # zmiana na maksymalizację dla komputera
          plansza[w][k] = " "

          if wynik < najlepsza_wartość:
            najlepsza_wartość = wynik
            najlepszy_ruch = (w, k)

    return najlepsza_wartość, najlepszy_ruch
```

```python
plansza = tworzenie_planszy()

while znak_gracz != "X" and znak_gracz != "O":
  znak_gracz = input("Wybierz X lub O: ").upper()

znak_komputer = "O" if znak_gracz == "X" else "X"

while tura != "T" and tura != "N":
  tura = input("Czy chcesz zaczynać? (T/N): ").upper()

tura = "gracz" if tura == "T" else "komputer"
```

```python
wyświetlenie_planszy(plansza)
while koniec != True and remis != True:

  if tura == "gracz":
    print("\nRuch gracza.")
    while pozycja != True:
      pozycja_kolumna = int(input("Podaj kolumnę (1-3):"))-1
      pozycja_wiersz = int(input("Podaj wiersz (1-3):"))-1
      pozycja = czy_poprawna_pozycja(plansza, pozycja_kolumna, pozycja_wiersz)

    plansza[pozycja_wiersz][pozycja_kolumna] = znak_gracz
    wyświetlenie_planszy(plansza)

    pozycja = False
    remis = czy_remis(plansza)
    koniec = czy_koniec(plansza)

  else:
    print("\nRuch komputera.")
    wiersz_komputer, kolumna_komputer = minmax(plansza, True)[1]
    plansza[wiersz_komputer][kolumna_komputer] = znak_komputer

    wyświetlenie_planszy(plansza)
    remis = czy_remis(plansza)
    koniec = czy_koniec(plansza)

  tura = "komputer" if tura == "gracz" else "gracz"


if remis == True:
  print("Remis! Brak wolnych pól.")
else:
  if tura == "gracz":
    print("Wygrał komputer!")
  else:
    print("Wygrał gracz!")
```
